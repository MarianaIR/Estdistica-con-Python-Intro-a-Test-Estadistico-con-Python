###**Visualizando y limpiando Movilens**

```python
# 1°) Hacemos mismo analisis que anteriormente con tmdb pero ahora con RATINGS.CSV que son las notas de Movilens de su dataset,
# cada linea NO es una pelicula es una evalucion del usuario, entonces creamos una VARIABLE NOTAS y le atribuimos el dataset --->

notas = pd.read_csv("ratings.csv")
notas.head()
```

```python
# 2°) Tenemos que AGRUPAR el conjunto de datos por MOVIEID y calcular el promedio(MEDIA) de RATING de la nota ---->

nota_promedio_por_pelicula = notas.groupby("movieId").mean()["rating"]
                               # agrupame           media   solo de rating
nota_promedio_por_pelicula.head()
# Nos devuelve promedio de nota(3.920930) de la pelicula de la ID 1 por ejemplo.
```

```python
# 3°) Grfaicamos para ver la Distribucion de esos datos, como es una SERIES le pasamos VALUES para que nos tome los datos --->

sns.distplot(nota_promedio_por_pelicula.values)

# A simple vista NO parece una Distribucion Normal, NO tenemos nota 0 pero si tenemos promedio con nota 5.
```

```python
# 4°) Hacemos un ejercicio de limpieza para excluir o limpiar las peliculas que tengan menos de 10 evaluaciones, y asi tener un
# grafico que se aproxime mas a la realidad, primero debemos contar cuantas notas tiene cada pelicula --->

cantidad_votos_por_pelicula = notas.groupby("movieId").count() #count=para contar votos
cantidad_votos_por_pelicula.head()
```

```python
# 5°) Ahora hacemos una QUERY para excluir los registros que tengan menos de 10 en puntuacion(RATING) --->

peliculas_10_o_mas_votos = cantidad_votos_por_pelicula.query("rating >= 10").index
                                                    #consulta, votos, index= porque queremos el INDICE de MovieId
peliculas_10_o_mas_votos

# Nos devuelve una SERIES, que se creo con cada uno de los ID que tienen 10 O MAS VOTOS.
```

```python
# 6°) Hacemos otra QUERY de la NOTA PROMEDIO por pelicula y usamos la funcion .LOC, localizamos la nota promedio
# por pelicula con MAS DE 10 VOTOS --->

nota_promedio_por_pelicula_10_o_mas_votos = nota_promedio_por_pelicula.loc[peliculas_10_o_mas_votos.values] # Localizame
nota_promedio_por_pelicula_10_o_mas_votos.head()

# Fueron excluidos los registro con pelicuela que tengan menos de 10 votos.
```

```python
# 7°) Graficamos ahora con los nuevos codigos --->

ax = sns.distplot(nota_promedio_por_pelicula_10_o_mas_votos.values)
ax.set(xlabel = "Nota Promedio", ylabel = "Densidad")
ax.set_title("Grafico de Distribucion del promedio de notas de Movilens")

# Tenemos el grafico limpio diferente al graficado anteriormente.
```

```python
# 8°) Hacemos el grafico sin la linea de Densidad y con la FRECUENCIA REAL y NO la normal --->

ax = sns.distplot(nota_promedio_por_pelicula_10_o_mas_votos.values, kde = False, norm_hist = False)
ax.set(xlabel = "Nota Promedio", ylabel = "Frecuencia")
ax.set_title("Grafico de Distribucion del promedio de notas de Movilens")
```

```python
# 9°) Graficamos BOXPLOT --->

ax = sns.boxplot(x = nota_promedio_por_pelicula_10_o_mas_votos.values)
ax.set(xlabel = "Nota Promedio")
ax.set_title("Grafico de Distribucion del Promedio de notas de Movilens")

# Nos muestra que el Q1 es mayor al Q3, Q2=Mediana

# RTA = PODEMOS DECIR QUE TMDB Y MOVILENS LOS GRAFICOS TIENEN UN COMPORTAMIENTO SIMILAR EN CUANTO A LOS VOTOS DE USUARIOS.
```

```python
# 2°) Calculamos la MEDIA de la MEDIA par analizar los promedios de esas notas --->

nota_promedio_por_pelicula_10_o_mas_votos.mean()

# Primero agrupamos el promedio de cada pelicula y luego calculamos el promedio del promedio de las peliculas
```

```python
# 5°) Vamos a calcular con un FOR (loop) para las 2269 peliculas, para calcular el promedio(MEDIA) del primero con el segundo,
# luego el primero con el segundo y el tercero, y asi sucesivamante para ver como se comprta el ACUMULADO de los promedios a
# medida que vamos avanzando con los datos ---->

promedios = list() # Creamos una variable PROMEDIOS que va a ser igual a una LISTA
for i in range(1, len(nota_promedio_por_pelicula_10_o_mas_votos)): # Tomamos el 1 resgitro de pelicula porque el anterior es 0 y analizamos los 2269 peliculas
  promedios.append(nota_promedio_por_pelicula_10_o_mas_votos[0:i].mean())#[0:i]= va ir a 1, luego 2, luego 3, etc y va a ir calculando el promedio

promedios

# Nos devuelvo el promedio el 1 con el segundo y asi sucesivamente de los 2269 registros
```

```python
# 6°) Graficamos esos promedios, antes importamos la bibilioteca MATPLOTLIB --->

import matplotlib.pyplot as plt

promedios = list() # Creamos una variable PROMEDIOS que va a ser igual a una LISTA
for i in range(1, len(nota_promedio_por_pelicula_10_o_mas_votos)): # Tomamos el 1 resgitro de pelicula porque el anterior es 0 y analizamos los 2269 peliculas
  promedios.append(nota_promedio_por_pelicula_10_o_mas_votos[0:i].mean())#[0:i]= va ir a 1, luego 2, luego 3, etc y va a ir calculando el promedio

promedios

plt.plot(promedios)

# Nos devuelve el grafico que comienza con 3.9, luego baja y sube muchos menos, los primeros datos muestran una OSCILACION muy grande
# lo que nos da la conclucion de si nuestra MUESTRA es PEQUEÑA siempre tenemos la probabilidad de tomar concluciones erradas, ya que
# podemos observar que a partir del numero 1000 se comienza a ESTABILIZAR.
```

```python
# 7°) Ahora vemos si el comportamiento del cuadro es el mismo cuando le damos un ORDEN ALEATORIO a nuestros datos o si cambia algo
# usamos ademas la bibilioteca NUMPY para crear ALEATOREAMENTE en nuestro arreglo --->

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(75243) # Colocamos una SEMILLA a nuestro RANDOM para que cada vez que ejecutemos NO nos genere un nuevo valor ALEATORIO diferente.
# (75243)= numero que probo el profesor
temp = nota_promedio_por_pelicula_10_o_mas_votos.sample(frac=1) # Creamos variable TEMP y que lo ordene con parametro FRAC

promedios = list()
for i in range(1, len(temp)): # Pasamos la variable nueva creada TEMP
  promedios.append(temp[0:i].mean())#[0:i]= va ir a 1, luego 2, luego 3, etc y va a ir calculando el promedio

promedios

plt.plot(promedios)

# Lo que hizo fue ordenar nuestro conjunto de datos, y podemos ver que comienza al inicio con una OSCILACION bien grande, pero que esta vez se
# ESTABILIZA en los 500. Entonces ademas del numero de MUESTRAS es importante el ORDEN tambien ya que puede INFLUENCIAR, por ejemplo si mis datos
# ya lo recibi ordenados me menos a mayor o visceversa, la probabilidad de una ERROR es mas significativa que con una muestra ordenada ALEATORIAMENTE
# y por ORDEN, DEBEMOS TOMAR EN CUENTA EL ORDEN!!!
```

###Intervalo de Confianza

```python
# 1°) Como podemos asegurar que los DATOS que usamos representan el universo de esos DATOS, para eso se utiliza los INTERVALOS
# DE CONFIANZA a traves de TEST ESTADISTICOS, se de un rango (range) de valores con una acertividad de estadistica donde el intervalo
# de confianza se encuentra en una MUESTRA de x e y, donde se encuentra la MEDIA, como vimos en NOTA que es un Distribucion Normal
# vamos a utilizar estadistica propia de ea Distribucion.
# Solo para refrescar lo media que logramos anteriormente en NOTA --->

nota_promedio_por_pelicula_10_o_mas_votos.mean()

# Esta MEDIA de voto de mas de 10 por pelicula, NO representa el universo de todos los datos
```

###***Z TEST - T TEST :***

```python
# 2°) Analizamos con TEST ESTADISTICOS, Z TEST Y T TEST, usamos modulo STATSMODELS y dentro de ese modulo tenemos
# STAST(estadistica), ademas utilizamos WIGHTSTATS y usamos ZCONFINT que es la Funcion para calcular el INTERVALO DE
# CONFIANZA en Z TEST ---->

# Z TETS :

from statsmodels.stats.weightstats import zconfint # del modulo importamos funcion zconfint

zconfint(nota_promedio_por_pelicula_10_o_mas_votos) # Pasamos como parametro de la Funcion la Variable de las notas.

# Nos devuelve el INTERVALO DE CONFIANZA, existe un 95% de Confianza de que si yo selecciono cualquier dato
# del Universo de este que este dentro de este INTERVALO DE CONFIANZA(3.41, 3.45)
```

```python
# 3°) Hacemos lo mismo con T TEST, pero se recomienda usar Z TEST con volumene grandes de datos y distribucion Normal,
# y T TEST con numero de datos mas pequeños de muestra, en T TEST NO tenemos una funcion especifica, lo que tenemos que
# hacer es importar en este modulo la Funcion DESCRSTATSW, para esto creamos una Varviable ---->

# T TEST :

from statsmodels.stats.weightstats import DescrStatsW

descr_notas_10_votos = DescrStatsW(nota_promedio_por_pelicula_10_o_mas_votos)
#variable              funcion     parametro, nuestra variable
descr_notas_10_votos.tconfint_mean() # metodo tconfint que esta dentro de DescrStatsW

# Nos devuelve el INTERVALO DE CONFIANZA de T TEST, parecido a Z TEST pero NO son iguales.
```

```python
# El MODULO DESCRSTATSW ademas de tener TCONFINT_MEAN() tiene ZCONFINT_MEAN()--->

descr_notas_10_votos.zconfint_mean()
```

```python
# 2°) Para asegurar esto dicho en los graficos existe un TEST para saber si son NORMALES o NO las Distribuciones, se llama
# NORMLTETS y esta dentro del modulo de SCIPY, nos devuelve 2 variables con valor P y un valor ESTADISTICO --->

from scipy.stats import normaltest

stat, p = normaltest(notas1.rating)
p

# P es menor a 0.05, en este caso RECHAZO la HIPOTESIS NULA = la distribucion es NORMAL, EN ESTE CASO NO ES NORMAL, por esto NO podemos
# utilizar ZTETS y TTETS, para esto existe el TEST de WILCOXON que usa datos NO PARAMETRICOS de Distribuiones que NO son Normales.
```

```python
# 3°) Usamos TEST DE WILCOXON, que no se basa en las MEDIAS de los valores, sino que compara las Distribuciones y los valores
# ALEATORIOS  de esas Distribuiones , utilizamos la funcion RANKSUMS, que devuelve un valor estadistico y un valor P ---->

from scipy.stats import ranksums

stat, p = ranksums(notas1.rating, notas593.rating) # Comparamos notas TS y SILENCE de MOVIELENS(NO NORMAL)
p

# Ho = Las 2 Muestras tienen la misma Distribucion(nuestra hipotesis nula)
# H1 = Los valores de una Muestra tienden a ser mayores o menores una de la otra(hipotesis alterna, contraria a Ho)

# Como PAVLUE < ALPHA (0.0003267718756440693 < 0.05) ---> NO Ho! ya que las muestras NO SON IGUALES!!
# SI H1, SILENCE OF THE LAMBS los valores son mayores a TOY STORY!
```
